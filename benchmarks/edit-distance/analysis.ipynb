```python
from pathlib import Path
import numpy as np
from matplotlib import pyplot as pl
%matplotlib notebook
```

```python
results = {}
root = Path('./results')
for r in Path('./results').glob('*/samples'):
    name = str(r.relative_to(root).parent)
    results[name] = np.genfromtxt(r)
```

```python
results
```

```python
pl.clf()
xlim = (0, max(max(*v) for v in results.values()))
for k,v in sorted(results.items()):
    v = v[10:-10]
    color = 'r' if k.startswith('no') else 'b'
    pl.hist(v, range=xlim, bins=100, alpha=0.4, label=k, color=color, log=True)
    
pl.legend()
```

```python
def stats(x):
    print('median', np.median(x))
    print('mean', np.mean(x))
    print('min', np.min(x))
    print('max', np.max(x))
```

```python
stats(results['no-90'])
```
