```python
import matplotlib.pyplot as pl
import numpy as np
```

```python
def read_times(fname): 
    queries = []
    adds = []
    for line in open(fname):
        if line.startswith('?'):
            queries.append(float(line[2:]))
        elif line.startswith('+'):
            adds.append(float(line[2:]))
    
    return {'queries': queries, 'additions': adds}
```

```python
no = read_times('no')
xn = read_times('xn')
```

```python
def hist_times(xn, no, bins=400, log=True, xmax=None):
    real_xmax = max(np.max(xn), np.max(no))
    if xmax is None:
        xmax = real_xmax
    hist_args = {
        'histtype': 'step',
        'alpha': 0.6,
        'bins': bins,
        'log': log,
        'cumulative': -1,
        'normed': True,
        'range': (0, real_xmax)
    }
    pl.hist(no, label='no', color='C1', **hist_args)
    pl.hist(xn, label='xn', color='C0', **hist_args)
    pl.axvline(np.mean(xn), label='xn', lw=0.5, color='C0')
    pl.axvline(np.mean(no), label='no', lw=0.5, color='C1')
    pl.xlim(0, xmax)
    pl.xlabel('latency (seconds)')
    pl.ylabel('inverse cumulative probability')
    print('xn', np.mean(xn))
    print('no', np.mean(no))
    pl.legend()
    pl.xlim(0,xmax)
```

```python
pl.figure()
hist_times(xn['queries'], no['queries'], xmax=0.5, bins=1000)
pl.suptitle('far-in query latency distribution')
```

```python
pl.figure()
hist_times(xn['queries'], no['queries'])
pl.suptitle('overall query latency distribution')
```

```python
pl.figure()
hist_times(xn['additions'], no['additions'])
pl.suptitle('overall index latency distribution')
```
